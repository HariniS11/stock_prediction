```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway

# Define categorical independent variables and continuous target
categorical_columns = ['Company', 'Ticker']
target = 'Close'

# Perform ANOVA for each categorical feature
anova_results = {}

for col in categorical_columns:
    groups = [df_1[df_1[col] == category][target] for category in df_1[col].unique()]
    f_stat, p_value = f_oneway(*groups)
    anova_results[col] = {'F-Statistic': f_stat, 'p-value': p_value}

# Convert results to DataFrame
anova_df_1 = pd.DataFrame(anova_results).T
print(anova_df_1.sort_values(by="F-Statistic", ascending=False))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x = df_1.drop('Close', axis=1)
y = df_1['Close']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.30, random_state=42)
```

```python
print(df_1.shape)
print('finding 70% =', (df_1.shape[0]/100) * 70)
print("X_train",x_train.shape)
print("y_train",y_train.shape)
print("X_test",x_test.shape)
print("y_test",y_test.shape)
```

```python
regression_model = LinearRegression()
regression_model.fit(x_train, y_train)
```

```python
y_predicted = regression_model.predict(x_test)
print("Testing data Mean Squared Error = ", mean_squared_error(y_test, y_predicted))
print("testing - R2 Score = ", r2_score(y_test, y_predicted))

y_predicted_train = regression_model.predict(x_train)
print("Training data Mean Squared Error = ", mean_squared_error(y_train, y_predicted_train))
print("training - R2 Score = ", r2_score(y_train, y_predicted_train))
```

# scaling and pickling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit the scaler on the raw data (before any scaling)
scaler.fit(df[['Open', 'High', 'Low', 'Volume']])

# Now the scaler has "learned" the mean and std dev of the raw data.
```

```python
# Transform the data using the fitted scaler
df[['Open', 'High', 'Low', 'Volume']] = scaler.transform(df[['Open','High', 'Low', 'Volume']])

# Save the fitted scaler to a pickle file
import pickle
with open("scaler_file.pkl", "wb") as f:
    pickle.dump(scaler, f)

print("Scaler saved successfully!")
```

```python
# Load the saved scaler from the pickle file
with open("scaler_file.pkl", "rb") as f:
    loaded_scaler = pickle.load(f)

# New raw data you want to transform (not scaled yet)
new_data = [[360, 355, 345, 5000000]]  # Raw data: [Open, Close, High, Low, Volume]

# Transform the new data using the loaded scaler
scaled_new_data = loaded_scaler.transform(new_data)

# Output the scaled data
print("Scaled New Data:", scaled_new_data)
```
