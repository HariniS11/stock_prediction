```python
import pandas as pd
df = pd.read_csv('FAANG_cleaned.csv')
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df = df.drop(['Analyst Recommendation', 'Ticker', 'Market Cap','PE Ratio','Beta','Forward PE','Net Income','Enterprise Value'], axis=1)
df = df.drop(['Debt to Equity','Return on Equity (ROE)','Current Ratio','Dividends Paid','Dividend Yield','Free Cash Flow'] , axis=1)
df = df.drop(['Profit Margin','Price to Book Ratio','Total Debt','Beta (5Y)','Annual Dividend Rate','Quick Ratio'] , axis=1)
df = df.drop(['Date','EPS','Quarterly Revenue Growth','Target Price','Operating Margin','Adj Close'], axis=1)
df = df.drop(10948, axis=0)
```

```python
lower_boundary = -172.95254989999998
upper_boundary = 319.53152746
df['Open'] = df['Open'].clip(upper = upper_boundary)

lower_boundary = -175.16379022499999
upper_boundary = 323.57827901499996
df['High'] = df['High'].clip(lower = lower_boundary , upper = upper_boundary)

lower_boundary = -171.2054237625
upper_boundary = 316.24324929750003
df['Low'] = df['Low'].clip(lower = lower_boundary , upper = upper_boundary)

lower_boundary = -173.152680125
upper_boundary = 319.843609995
df['Close'] = df['Close'].clip(lower = lower_boundary , upper = upper_boundary)
```

```python
lower_boundary = -142146900.0
upper_boundary = 293751500.0
df['Volume'] = df['Volume'].clip(lower = lower_boundary , upper = upper_boundary)
```

```python
df[df['Company'] == 'Amazon']
```

```python
import pickle
from sklearn.preprocessing import StandardScaler

with open("scaler_file.pkl", "rb") as f:
    loaded_scaler = pickle.load(f)

# New raw data you want to transform (not scaled yet)
scaler = StandardScaler()
df[['Open', 'High', 'Low', 'Volume']] = scaler.fit_transform(df[['Open', 'High', 'Low', 'Volume']])
```

```python
scaler = StandardScaler()
df[['Close']] = scaler.fit_transform(df[['Close']])
```

```python
import pandas as pd
df = pd.get_dummies(df, columns=['Company'])
df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x = df.drop('Close', axis=1)
y = df['Close']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.20, random_state=42)
```

```python
print(df.shape)
print('finding 70% =', (df.shape[0]/100) * 70)
print("X_train",x_train.shape)
print("y_train",y_train.shape)
print("X_test",x_test.shape)
print("y_test",y_test.shape)
```

```python
regression_model = LinearRegression()
regression_model.fit(x_train, y_train)
```

```python
y_predicted_train = regression_model.predict(x_train)
y_predicted = regression_model.predict(x_test)

mse = mean_squared_error(y_test, y_predicted)
r2 =  r2_score(y_test, y_predicted)

print("Testing data Mean Squared Error = ", mse)
print("testing - R2 Score = ", r2)
```

Decision Tree

```python
# FINDING BEST PARAMETERS
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
dt = DecisionTreeRegressor()
```

```python
parameters = {
    'max_depth': [3, 5, 10, 15, None],       
    'min_samples_split':[2,4,6,8],
    'min_samples_leaf':[1,2,3,4,5] ,        
    }
```

```python
# Apply GridSearchCV
grid_search = GridSearchCV(estimator=dt, param_grid = parameters,n_jobs = -1)  
```

```python
grid_search.fit(x_train, y_train)
best_params = grid_search.best_params_
print("Best Parameters:", best_params)
```

```python
best_params = {'max_depth': None, 'min_samples_leaf': 5, 'min_samples_split': 8}
```

```python
dt = DecisionTreeRegressor(**best_params)
X_train,  X_test, y_train, y_test = train_test_split(x, y, test_size= 0.30, random_state=42)
dt.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
y_predicted = dt.predict(X_test)

r2 = r2_score(y_test,y_predicted)
mse = mean_squared_error(y_test,y_predicted)
rmse = np.sqrt(mean_squared_error(y_test,y_predicted))
print('mean square error',mse)
print('r2 score',r2)
print('rmse', rmse)
```
